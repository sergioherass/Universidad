# MatPlotLib 

El paquete *matplotlib* sirve para dibujar o representar gráficamente todo tipo de cosas (gráficas de funciones, estadísticas, puntos en un plano, fractales, dibujos, etc.)

Es muy versátil. Aquí veremos sólo unos pequeños ejemplos que servirán sólo para entender su funcionamiento básico y apreciar su potencial.

Cargamos el paquete *matplotlib* como *plt*

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import math
```

### Dibujo de una función
Vamos a dibujar una función. Para ello primero creamos los puntos sobre el eje $X$. Aquí lo haremos con una separación de $0.05$.

```python
x = np.arange(0, math.pi*2, 0.05) 
```

Observa que hemos usado $\pi = $`math.pi`. Ahora de cada punto calculamos sus imágenes. En lugar de hacer un `for` o de hacer una lista comprimida `[np.sin(t) for t in x]`, podemos aplicar directamente `np.sin()` sobre el array.

```python
y = np.sin(x)
```

Y ahora creamos la imagen (o el dibujo):

```python
plt.plot(x,y)# Con esto dibujamos la función f(x)=sin(x)
plt.show()# Mostramos el dibujo
```

Es posible mejorar nuestro dibujo añadiendo etiquetas al eje $X$, al eje $Y$ y un título a nuestro dibujo. Esto se hace con `plt.xlabel`, `plt.ylabel` y `plt.title`.

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle") # Añadimos la etiqueta del eje X
plt.ylabel("sine") # Añadimos la etiqueta del eje Y
plt.title('sine wave') # Añadimos el título a nuestra gráfica
plt.show()
```

### Dibujo de una función lineal a trozos
También podemos una gráfica lineal a trozos, debido a que pueden ser datos. Imaginemos que tenemos siguientes datos

```python
# import matplotlib.pyplot as plt
y = [1, 4, 9, 16, 25, 36,49, 64]
x1 = [1, 16, 30, 42, 55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
```

Aquí tenemos dos variables: la variable $y$, que representa las ventas conseguidas, y  la variable $X$, que mide (de alguna manera) la publicidad (en alguna medida desconocida).

Ahora realizamos el dibujo de la gráfica a través de los datos. 
1. Primero creamos la figura con `plt.figure()` (ver https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.figure.html para detalles): lo guardamos en la variable `fig`; 
2. luego crearemos los ejes con un método de la figura `fig.add_axes` (las entradas son las dimensiones de los ejes), que genera otra variable que almacenamos en `ax`.
3. y añadimos los datos usando el método `ax.plot`. Entre las diferentes posibilidades está el poder poner color, (`y` amarillo, `g` verde, `b` azul, etc.), el trazado (`-` línea continua, `--` línea discontinua), y los puntos (`s` para marcadores cuadrados, `o` para marcadores redondos).
4. Luego le añadimos todo tipo de detalles como la leyenda (`ax.legend`), el título (`ax.set_title`), y las etiquetas de los ejes (`ax.set_xlabel`).
5. Finalmente con `plt.show()` creamos el gráfico.



```python
# import matplotlib.pyplot as plt
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'ys-') # línea continua (-) amarilla (aka. *y*ellow) con marcadores cuadrados (*s*quare)
l2 = ax.plot(x2,y,'go--') # línea discontinua (--) de color verde (*g*reen) con marcadores redondos (o)
ax.legend(labels = ('Televisores', 'Móviles'), loc = 'lower right') # Leyenda situada en la esquina inferior derecha
ax.set_title("Efecto de los anuncios en las ventas")
ax.set_xlabel('Publicidad')
ax.set_ylabel('Ventas')
plt.show()
```

### Diagramas

También podéis hacer diagrama de barras de datos que poseéis. He aquí un ejemplo.

```python
# import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']# Cada uno de los datos
students = [23,17,35,29,12]# Datos (¿Cómo es posible que Java tenga tantos alumnos?)
ax.bar(langs,students)
plt.show()
```

También podemos hacer una gráfica de sectores

```python
#from matplotlib import pyplot as plt
#import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students, labels = langs,autopct='%1.2f%%')# en lugar de ax.bar, usamos ax.pie
# %1.2f%% es el formato con el que queremos que aparezcan los datos dentro de los sectores 
plt.show()
```

### Trazado de puntos, rectas y polígonos

También podemos dibujar puntos, líneas y figuras.

Para dibujar tres puntos en un plano usamos `plt.scatter`. En este caso, la variable `x` tiene las primeras coordenadas de todos los puntos mientras que `y` tiene las segundas coordenadas de los puntos.

```python
# from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = np.array([0,1,2])
y = np.array([1,2,0])
plt.scatter(x,y)# Dibujamos tres puntos (0,1), (1,2) y (2,0)
plt.show()
```

Para dibujar líneas usamos `plt.plot()`. En este caso almacenamos los puntos como siguen: una recta está definida por dos puntos $(x_1,y_1)$ y $(x_2,y_2)$. Para dibujar la recta, creamos dos variables 
***
x1 = np.array([$x_1,x_2 $]),

y1 = np.array([$y_1,y_2$]).
***
y con `pl.plot(x1,y1)` dibujaríamos la recta. En el siguiente ejemplo, hemos dibujado dos rectas. Observa que por defecto asigna a cada recta un color distinto.

```python
# from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# Puntos iniciales: (0,1) y (1,0)
x1 = np.array([0,1])
y1 = np.array([1,0])
# Puntos finales: (1,-1) y (-1,1)
x2 = np.array([1,-1])
y2 = np.array([-1,1])
plt.plot(x1, y1, x2, y2, marker='o')# Dibujamos las dos líneas que parten de los puntos de partida hasta los puntos finales 
# marker = 'o' sirve para dibujar lo que marca
plt.show()
```

Para dibujar un polígono de vértices $p_i=(x_i,y_i)$ tenemos que crear un array con todos los puntos como por ejemplo
***
`edges = np.array([`$p_1,p_2,\ldots,p_r$`])`
***
Que luego con el comando `plt.Polygon` generamos el polígono:
***
`polygon = plt.Polygon(edges, color = 'red', alpha = 0.2)`
***
Aquí podemos poner color y transparencia (con `alpha`). Añadimos a nuestro dibujo la figura:

```python
#import numpy as np 
#import matplotlib.pyplot as plt

edges1 = np.array([[1.,0.],[0.,1.],[0.,2.]])

plt.figure()

fig1 = plt.Polygon(edges1, color='blue',alpha=0.1)
plt.gca().add_patch(fig1)

edges2 = np.array([[0.,0.],[0.,0.5],[0.5,0.5]])

fig2 = plt.Polygon(edges2, color='red',alpha=0.2)
plt.gca().add_patch(fig2)

plt.show()
```
