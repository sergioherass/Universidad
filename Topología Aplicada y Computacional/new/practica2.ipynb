```python
import matplotlib.pyplot as plt
import numpy as np

# Lista de puntos (coordenadas x, coordenadas y)
puntos = [(0.0, 0.32475952617543313), (0.0, 0.32475952617543313), (0.0, 0.4100304866714181), (0.0, 0.4100304866714181), (0.0, 0.4685926655361565), (0.5000000000933408, 0.5159301380161246), (0.5000000000933408, 0.5159301380161246), (0.5, 0.5173076923076921), (0.4685926655361565, 0.5773502692614795)]

# Extraer las coordenadas x e y de la lista de puntos
x_coords, y_coords = zip(*puntos)

# Encontrar la coordenada y máxima
max_y = np.ceil(max(y_coords))
print(max_y)
# Plotear los puntos
plt.scatter(x_coords, y_coords, color='blue', marker='o')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')

# Establecer límites de ejes redondeando al alza la coordenada y máxima
plt.xlim(0, max_y)
plt.ylim(0, max_y)

plt.title('Plot de una Lista de Puntos')
plt.show()

```
