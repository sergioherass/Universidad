```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los semiejes de la elipse
a = 2.0
b = 1.0

# Número de puntos
num_points = 35

# Calcular las coordenadas paramétricas de los puntos con un componente aleatorio
theta = np.linspace(0, 2*np.pi, num_points)
x = a * np.cos(theta) + 0.1 * np.random.normal(size=num_points)
y = b * np.sin(theta) + 0.1 * np.random.normal(size=num_points)

# Crear una lista de puntos
points = list(zip(x, y))

# Graficar los puntos
plt.plot(x, y, 'o', label='Elipse Aproximada')

# Configurar el aspecto de la gráfica
plt.axis('equal')  # Asegurar que los ejes tengan la misma escala
plt.title('Aproximación de una Elipse con 50 Puntos Desplazados')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()

# Imprimir la lista de puntos
print("Lista de Puntos:")
for point in points:
    print(point)
```
