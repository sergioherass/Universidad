```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.tri import Triangulation

# Parámetros de la superficie de género dos
theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, 2 * np.pi, 100)
theta, phi = np.meshgrid(theta, phi)
r = 1  # Radio de los toros
d = 2  # Distancia entre los toros

# Coordenadas paramétricas de la superficie de género dos
x = (r * np.cos(theta) + d) * np.cos(phi)
y = (r * np.cos(theta) + d) * np.sin(phi)
z = r * np.sin(theta)

# Crear la figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Triangulación
tri = Triangulation(np.ravel(x), np.ravel(y))

# Visualización de la triangulación
ax.plot_trisurf(tri, np.ravel(z), cmap='viridis', edgecolor='k')

# Ajustes de la visualización
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Triangulación del Doble Toro')

plt.show()
```
